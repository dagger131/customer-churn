```python
def conv(value):
    if value==1:
        return "yes"
    else:
        return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

so here what have we done ?

senior citizen was giving two values 0 or 1 for no or yes resp., so here we have converted 0 to no and 1 to yes

EXPLORATION OF DATA

```python
ax  = sns.countplot(x= 'Churn', data = df)
ax.bar_label(ax.containers[0])
plt.title("Count of customer by churn ")
plt.show()
```

```python
gb = df.groupby("Churn").agg({'Churn' : "count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%" )
plt.title("Percentage of Churned Customers")
plt.show()
```

from the given pie chart we can concludethat 26.54 % of our customers have churned out. Now let's explore the reason behind it.

on the basis of gender

```python
plt.figure(figsize = (3,3))
ax = sns.countplot(x= df["gender"], data = df, hue =  "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.show()
```

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x= df["SeniorCitizen"], data = df, hue =  "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.show()
```

comparatively, a greater percentage of people in senior citizen have churned.

on the basis of tenure

```python
plt.figure(figsize = (9,4))
sns.histplot(x = "tenure",data = df, bins = 72, hue= 'Churn')
plt.show()
```

 people who stayed with us for a longer period of time have not churned but most of the people/ customer churned within first few months.
 

```python
plt.figure(figsize = (4,4))
ax = sns.countplot(x= df["Contract"], data = df, hue =  "Churn")
ax.bar_label(ax.containers[0])
ax.bar_label(ax.containers[1])
plt.show()
```

people whohave month to month contract are likely to churn then from those who have 1 or 2 years of contract

```python
# Assuming df is your DataFrame
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Determine the layout of subplots
n_cols = 3
n_rows = -(-len(columns) // n_cols)  # Ceiling division to arrange all columns

# Create subplots
fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 5 * n_rows))
axes = axes.flatten()

# Generate countplots for each column
for i, col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i], hue = df['Churn'])
    axes[i].set_title(f"Countplot of {col}")
    axes[i].set_xlabel('')
    axes[i].set_ylabel('Count')

# Remove empty subplots if any
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Adjust layout
plt.tight_layout()
plt.show()
```

Most customers opt for basic phone service and DSL internet, with fewer choosing fiber optics.
Services like online security, tech support, and device protection show higher churn rates among those who do not use them.
"No internet service" consistently corresponds to minimal churn, suggesting stable users in that group.
